```python
def outlier_var(x):
    if ((x.dtype=='float') or (x.dtype=='int')):
        x= x.clip(lower = x.quantile(0.01), upper = x.quantile(0.99))
    else:
        x
    return x
```

```python
df = df.apply(outlier_var)
```

```python
# List of variables to map

varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(binary_map)
```

```python
# get the significant variables : bi-variate regression
from sklearn.feature_selection import f_regression
```

```python
train, test = train_test_split(df_new, test_size=0.3, random_state=123)
```

```python
import statsmodels.formula.api as smf
```

```python
train.columns
```

```python
model_ols = smf.ols('ln_price ~ airconditioning +guestroom + parking + prefarea + stories+unfurnished', data=train).fit()
```

```python
print(model_ols.summary())
```

```python
#prediction of new house
airconditioning=1
guestroom=1
parking=1
prefarea=0
stories=0
unfurnished=1
```

```python
ln_price=14.8706+0.2343*airconditioning+0.1283*guestroom+0.0974*parking+0.1929*prefarea+0.1325*stories-0.1524*unfurnished
```

```python
import math
price = math.exp(ln_price)
price
```

```python
sns.distplot(model_ols.resid)
```

```python
train_pred = np.exp(model_ols.predict(train))
test_pred = np.exp(model_ols.predict(test))
train_act = np.exp(train.ln_price)
test_act = np.exp(test.ln_price)
```

```python
print('train_MAPE:', np.mean(np.abs(train_act- train_pred)/train_act))
```

```python
print('test_MAPE:', np.mean(np.abs(test_act- test_pred)/test_act))
```

```python
print('train_RMSE:', np.sqrt(np.mean((train_act- train_pred)**2)))
print('test_RMSE:', np.sqrt(np.mean((test_act- test_pred)**2)))
```

```python
print('train_corr:', np.corrcoef(train_act, train_pred)[1][0])
print('test_corr:', np.corrcoef(test_act, test_pred)[1][0])
```
